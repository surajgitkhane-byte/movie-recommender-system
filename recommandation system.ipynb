```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    if isinstance(text, str):
        # Handle strings: split, stem, and join
        y = []
        for i in text.split():
            y.append(ps.stem(i))
        return " ".join(y)
    elif isinstance(text, list):
        # Handle lists (assuming they contain strings): stem each, then join
        stemmed_list = []
        for item in text:
            if isinstance(item, str): # Ensure elements within the list are strings
                stemmed_list.append(ps.stem(item))
            else:
                # Handle non-string elements within the list (e.g., numbers, None)
                # You can choose to skip them, convert to string, or raise an error
                stemmed_list.append(str(item)) # Convert to string if not already
        return " ".join(stemmed_list)
    else:
        # Handle other types (e.g., NaN, NoneType, numbers directly in the column)
        # Convert them to string before processing, or return as is based on your need.
        return str(text) # Convert to string to avoid errors with other operations


new_df['tags'] = new_df['tags'].apply(stem)
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
   
```
